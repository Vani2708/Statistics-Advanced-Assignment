**Question 1: What is Computational Linguistics and how does it relate to NLP**?

Answer:
Computational Linguistics is the study of how computers can understand, analyze, and generate human language using linguistic rules and computational models.

It relates to Natural Language Processing (NLP) because NLP is the practical application of computational linguistics techniques to build systems such as chatbots, machine translation, and speech recognition tools.

**Question 2: Briefly describe the historical evolution of Natural Language Processing.**

Answer:
The development of NLP can be divided into several stages:

**1950s–1960s:** (Rule-Based Systems)
Early NLP systems relied on manually written grammar rules for tasks such as machine translation.

**1980s–1990s:** (Statistical Methods)
NLP began using probability and statistical models trained on large text datasets.

**2000s–Present:** (Machine Learning & Deep Learning)
Modern NLP uses machine learning, neural networks, and large language models to perform tasks like sentiment analysis, chatbots, and language translation.

**Question 4: What is text normalization and why is it essential in text processing tasks**?

Answer:
Text normalization is the process of transforming text into a consistent and standard format before analysis.

Common normalization steps include:

Converting text to lowercase

Removing punctuation

Removing stopwords

Expanding abbreviations

It is essential because it reduces variation in text data, making it easier for NLP models to process and analyze language effectively.

**Question 5: Compare and contrast stemming and lemmatization with examples.**

**Answer:**
**Stemming** is a text processing technique that removes word endings to produce a base form that may not always be a real word (e.g., *studies → studi*), whereas **lemmatization** converts a word to its correct dictionary base form using linguistic rules (e.g., *studies → study*), making it more accurate but computationally slower.



**Question 6: Python program to extract email addresses using regex**

```python
import re

text = """Hello team, please contact us at support@xyz.com for technical issues,
or reach out to our HR at hr@xyz.com. You can also connect with John at
john.doe@xyz.org and jenny via jenny_clarke126@mail.co.us.
For partnership inquiries, email partners@xyz.biz."""

pattern = r'[a-zA-Z0-9._%+-]+@[a-zA-Z0-9.-]+\.[a-zA-Z]+'

emails = re.findall(pattern, text)

print("Extracted Emails:")
for email in emails:
    print(email)
```

**Question 8: Custom annotator to identify proper nouns (spaCy)**

```python
import spacy

nlp = spacy.load("en_core_web_sm")

text = "John works at Google and lives in New York."

doc = nlp(text)

print("Proper Nouns:")
for token in doc:
    if token.pos_ == "PROPN":
        print(token.text)
```
