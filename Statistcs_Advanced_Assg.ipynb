**Question 1: What is a random variable in probability theory**?

Answer:
A random variable is a numerical value that represents the outcome of a random experiment. Instead of listing outcomes directly, we assign numbers to them so they can be analyzed mathematically.

For example, if we toss a coin:

Heads = 1

Tails = 0

The variable representing the result is a random variable because its value depends on chance.

**Question 2: What are the types of random variables**?

Answer:
There are two main types of random variables:

Discrete Random Variables

Take countable values.

Example: number of students in a class, number of coin toss successes.

Continuous Random Variables

Take any value within a range.

Example: height, weight, temperature, time.

**Question 3: Explain the difference between discrete and continuous distributions.**

Answer:
A discrete distribution represents probabilities of countable values such as the number of students in a class, while a continuous distribution represents probabilities of values within a continuous range such as height, weight, or time, where the variable can take infinitely many possible values within an interval.

**Question 5: What is the standard normal distribution, and why is it important**?

Answer:
The standard normal distribution is a special case of the normal distribution with:

Mean (μ) = 0

Standard deviation (σ) = 1

It is important because:

It allows comparison of different datasets using z-scores.

Many statistical methods rely on it.

It helps calculate probabilities using standard normal tables.

**Question 6: What is the Central Limit Theorem (CLT), and why is it critical in statistics**?

Answer:
The Central Limit Theorem (CLT) states that when the sample size is sufficiently large, the distribution of the sample mean approaches a normal distribution, regardless of the original population distribution.

Importance:

Allows normal approximation for many statistical methods.

Enables confidence intervals and hypothesis testing.

Makes analysis easier even when the population distribution is unknown.

**Question 7: What is the significance of confidence intervals in statistical analysis**?

Answer:
A confidence interval (CI) provides a range of values that likely contains the true population parameter.

Example:
A 95% confidence interval means we are 95% confident that the true population mean lies within the calculated range.

Importance:

Shows uncertainty in estimates.

More informative than a single value.

Widely used in decision-making and research.

Question 9: Write a Python program to generate 1000 random numbers from a normal
distribution with mean = 50 and standard deviation = 5. Compute its mean and standard
deviation using NumPy, and draw a histogram to visualize the distribution.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random numbers
data = np.random.normal(loc=50, scale=5, size=1000)

# Compute mean and standard deviation
mean = np.mean(data)
std_dev = np.std(data)

print("Mean:", mean)
print("Standard Deviation:", std_dev)

# Plot histogram
plt.hist(data, bins=30, color='skyblue', edgecolor='black')
plt.title("Histogram of Normal Distribution (mean=50, std=5)")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```

```python
import numpy as np
from scipy import stats

daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
               235, 260, 245, 250, 225, 270, 265, 255, 250, 260]

# Convert to numpy array
data = np.array(daily_sales)

# Calculate mean
mean_sales = np.mean(data)

# Standard error
std_error = stats.sem(data)

# 95% confidence interval
confidence = 0.95
ci = stats.t.interval(confidence, len(data)-1, loc=mean_sales, scale=std_error)

print("Mean Daily Sales:", mean_sales)
print("95% Confidence Interval:", ci)
```
